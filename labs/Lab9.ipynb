```python
import seaborn as sns
import matplotlib.pyplot as plt

print("\nБаланс выборки (ресэмплинг): ")
target_count = df.target.value_counts()
print(target_count)
print('Class 0: ', target_count[0])
print('Class 1: ', target_count[1])
print('Пропорция: ', round(target_count[0] / target_count[1], 2), ': 1')
sns.countplot(data=df, x='target')
plt.box(False)
plt.xlabel('target 1/0', fontsize=11)
plt.ylabel('Количество наблюдений', fontsize=11)
plt.show()
```

```python
shuffled_df = df.sample(frac=1, random_state=4)
CHD_df = shuffled_df.loc[shuffled_df['target'] == 1]
none_CHD_df = shuffled_df.loc[shuffled_df['target'] == 0].sample(n=250, random_state=42)
normalize_df = pd.concat([CHD_df, none_CHD_df])
target_count2 = normalize_df.target.value_counts()
print(target_count2)
sns.countplot(data=normalize_df, x='target')
plt.box(False)
plt.xlabel('target 1/0', fontsize=11)
plt.ylabel('Количество наблюдений', fontsize=11)
plt.show()
```

```python

X = df.iloc[:,:]
X = X.drop(['target'], axis=1)
X = X.drop(['oldpeak'], axis=1)
y = df['target']

from imblearn.under_sampling import TomekLinks

tl = TomekLinks(sampling_strategy='majority')
X_res, y_res = tl.fit_resample(X, y)
# print(X_res.shape)
# print(X.shape)
df_new = pd.concat([X_res, y_res], axis=1)
print(df_new)
sns.countplot(data=df_new, x=df_new.target)
plt.title('Under-sampling')
plt.box(False)
plt.xlabel('quality 1/0', fontsize=11)
plt.ylabel('Количество наблюдений', fontsize=11)
plt.show()
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X, y)
print(X_sm.shape)
print(X.shape)
df_new_sm = pd.concat([X_sm, y_sm], axis=1)
print(df_new_sm.shape)
sns.countplot(data=df_new_sm, x='target')
plt.title('Over-sampling')
plt.box(False)
plt.xlabel('target 1/0', fontsize=11)
plt.ylabel('Количество наблюдений', fontsize=11)
plt.show()
```
