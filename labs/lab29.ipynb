```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
df = pd.read_csv('heart.csv')
df = df.dropna()
df.head(10)
```

```python
df = df.ffill(axis=1)
df
```

```python
transactions = []
for i in range(0, 1000):
    transactions.append([str(df.values[i,j]) for j in range(0, 14)])
```

```python
result = list(apriori(transactions, min_support = 0.01, min_confidence = 0.1, min_lift = 4, min_length = 2))
```

```python
association_results = list(result)
```

```python
print(len(result))
```

```python
rules=[]
for item in result:

    # первый индекс внутреннего списка
    # Содержит базовый предмет и добавляет предмет
    pair = item[0]
    items = [x for x in pair]
    v1=str(items[0])
    v2= str(items[1])

    #второй индекс внутреннего списка
    v3= str(item[1])

    #третий индекс списка, расположенный на 0-м месте
    #третьего индекса внутреннего списка

    v4= str(item[2][0][2])
    v5=str(item[2][0][3])
    rows=(v1,v2,v3,v4,v5)
    rules.append(rows)
label=['title1','title2','suppot','confidence','lift']
nm=pd.DataFrame.from_records(rules,columns=label)
nm
```
